4.a.Write a user defined function for:  Titles of all the movies on the page as a Python list

```python
def get_movie_titles(soup_movie):
    movie_titles = []
    for movie in soup_movie.find_all("div", class_="card style_1"):
        title = movie.find("h2").text.strip()
        movie_titles.append(title)  # Append the title to the list
    return movie_titles

# Example:
movie_titles = get_movie_titles(soup_movie)
movie_titles
```

4.b. User ratings of all the movies on the page as a Python list

```python
def get_user_ratings(soup_movie):
    user_ratings = []
    for movie in soup_movie.find_all("div",class_ = "card style_1"):
      if movie.find('div', class_ ='user_score_chart') == 'null':
        user_ratings.append("not rated")
      else:
        user_ratings.append(movie.find('div', class_ ='user_score_chart').get("data-percent"))
    return user_ratings

# Example:
user_ratings = get_user_ratings(soup_movie)
user_ratings
```

4.C. HTML content of all the individual pages of movies collected into a Python list

```python
def get_movie_pages(soup_movie):
    movie_pages = []
    for movie in soup_movie.find_all("div",class_ = "card style_1"):
        movie_pages.append(movie.find('a').get("href"))
    return movie_pages

#Example
movie_pages = get_movie_pages(soup_movie)
movie_pages
```

4.d. Genres of all the movies on the page as a Python list

```python
def get_movie_generes(soup_movie):
  all_movie_generes = []
  movie_pages = get_movie_pages(soup_movie)
  for movie_page in movie_pages:
    url = "https://www.themoviedb.org" + movie_page
    needed_headers = {'User-Agent': "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36",'Accept-Language': 'en'}
    response = requests.get(url,headers = needed_headers) # Get the response
    response.raise_for_status() # Checking for HTTP errors
    movie_soup = BeautifulSoup(response.content, 'html.parser') # Pass the HTML content to BeautifulSoup
    #all generes
    genere_section = movie_soup.find('span', class_='genres')
    if genere_section:
         genere_list = genere_section.find_all('a')
         movie_genere = [genere_member.get('href').split('/')[2].split('-')[1] for genere_member in genere_list]
         all_movie_generes.append(movie_genere)
    else:
         all_movie_generes.append([])  # Append an empty list if no cast found
  return all_movie_generes

#Example
movie_generes = get_movie_generes(soup_movie)
print(movie_generes)
```

4.e. Cast of all the movies on the page as a Python list

```python
import requests
from bs4 import BeautifulSoup

def get_movie_cast(soup_movie):
    all_movie_cast = []
    movie_pages = get_movie_pages(soup_movie)
    for movie_page in movie_pages:
        url = "https://www.themoviedb.org" + movie_page
        needed_headers = {'User-Agent': "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36",'Accept-Language': 'en'}
        response = requests.get(url,headers = needed_headers)
        response.raise_for_status()
        movie_soup = BeautifulSoup(response.content, 'html.parser')

        # Find the cast section
        cast_section = movie_soup.find('ol', class_='people scroller')
        if cast_section:
            cast_list = cast_section.find_all('li', class_='card')
            movie_cast = [cast_member.find('p').text.strip() for cast_member in cast_list]
            all_movie_cast.append(movie_cast)
        else:
            all_movie_cast.append([])  # Append an empty list if no cast found

    return all_movie_cast

# Example)
movie_cast_list = get_movie_cast(soup_movie)
print(movie_cast_list)
```

6.a. Write a function that scrapes data (mentioned in Q5) from page number 1, 2, 3, 4 and 5 on
the URL https://www.themoviedb.org/movie and returns 5 data frames which can be exported
to csv file by calling the functions defined in Q3a, Q4c and Q5 (3 marks)
Hint : use the string “ ?page={ } “ concatenated to the string
“ https://www.themoviedb.org/movie “ which extracts the data from desired page no passed to
page value

```python
import pandas as pd

def scrape_movie_and_export():
    dataframes = []
    for page_number in range(1, 6):
        url = "https://www.themoviedb.org/movie" + f"?page={page_number}"
        soup_new = url_converter(url)  # Using your function from 2c
        title_df = get_movie_titles_df(soup_new)
        ratings_df = get_user_ratings_df(soup_new)
        genres_df = get_movie_generes_df(soup_new)
        cast_df = get_movie_cast_df(soup_new)

        # Combine the dataframes for this page
        pages_df = pd.concat([title_df, ratings_df, genres_df, cast_df], axis=1)
        dataframes.append(pages_df)

        # Optionally, export each page's dataframe to a CSV
        pages_df.to_csv(f"movie_data_page{page_number}.csv", index=False)

    return dataframes
```

```python
scrape_movie_and_export()
```

6.b. Combine the data obtained from dataframes in Q6

```python
# Call the function to get the list of dataframes
dataframes = scrape_movie_and_export()

# Combine all dataframes into a single dataframe
combined_df = pd.concat(dataframes, ignore_index=True)
print(combined_df)
```
